```r
## Installed "jpeg" and "raster" packages
```

```r
library("jpeg")
library("raster")
```

```r
##Reading image
```

```r
img=readJPEG("/Users/Lenovo/Desktop/ppm.jpg")
```

```r
##Structure of the image
```

```r
str(img)
```

```r
##Dimension of the image
```

```r
dim(img)
```

```r
##For displaying the image we use rasterImage
```

```r
par(mfrow=c(1,1))
plot(0:512, 0:512, type="n")
rasterImage(img, 0, 0, 512, 512,interpolate=FALSE)
```

```r
img_red=img
img_red[,,2:3]=0
plot(0:512, 0:512, type="n")
rasterImage(img_red, 0, 0, 512, 512,interpolate=FALSE)
img_green=img
img_green[,,c(1,3)]=0
plot(0:512, 0:512, type="n")
rasterImage(img_green, 0, 0, 512, 512,interpolate=FALSE)
img_blue=img
img_blue[,,1:2]=0
plot(0:512, 0:512, type="n")
rasterImage(img_blue, 0, 0, 512, 512,interpolate=FALSE)
```
